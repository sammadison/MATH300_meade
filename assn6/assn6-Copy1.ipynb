```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Lorenz(pt, a=10, b=28, c=8/3, dt=.01):
    (x,y,z) = pt
    x_0 = (a*(y-x))*dt
    y_0 = (x*(b-z) - y)*dt
    z_0 = (x*y - c*z)*dt
    return (x+x_0, y+y_0, z+z_0)
```

```python
u=(1,1,1)
Lorenz(u)
```

```python
m=(1,20,34)
Lorenz(m)
```

```python
n = 1000
pt = (1,1,1)
seq=np.empty((n,3))
```

```python
for i in range(n):
    seq[i]=pt
    pt=Lorenz(pt)
```

```python
seq
```

```python
plt.plot(seq[:,1], seq[:,2], 'r-')
plt.xlabel('Y-column')
plt.ylabel('Z-column')
plt.title('Lorenz Attractor')
plt.show()
```
